# Platform Compatibility Tests

> Tests to verify the package works across different platforms (Colab, Deepnote, Local)

```python
#| export
import sys
import importlib.util
```

## Core Import Tests

```python
def test_core_imports():
    """Test that all core modules can be imported"""
    import data401_nlp
    import data401_nlp.helpers.env
    import data401_nlp.helpers.llm
    import data401_nlp.helpers.submit
    import data401_nlp.helpers.spacy
    assert True
```

```python
def test_version():
    """Test that version is set correctly"""
    from data401_nlp import __version__
    assert __version__ is not None
    assert isinstance(__version__, str)
    assert len(__version__) > 0
```

## Environment Detection Tests

```python
def test_environment_detection():
    """Test that environment detection works"""
    from data401_nlp.helpers.env import load_env
    env = load_env()
    assert env in ["colab", "dotenv", "external"]
```

## Dependency Checks

```python
def test_core_dependencies():
    """Test that core dependencies are available"""
    required_packages = [
        'fastcore',
        'nbformat',
        'pandas',
        'orjson',
        'dotenv',  # python-dotenv imports as dotenv
        'IPython',
        'httpx',
    ]
    
    for package in required_packages:
        spec = importlib.util.find_spec(package)
        assert spec is not None, f"Required package '{package}' not found"
```

```python
def test_optional_nlp_dependencies():
    """Test if optional NLP dependencies are available (non-critical)"""
    optional_packages = ['spacy', 'nltk']
    available = {}
    
    for package in optional_packages:
        spec = importlib.util.find_spec(package)
        available[package] = spec is not None
    
    # Just log availability, don't fail
    print(f"NLP package availability: {available}")
    assert True  # Always pass
```

## Platform-Specific Tests

```python
def test_python_version():
    """Test that Python version meets minimum requirements"""
    assert sys.version_info >= (3, 11), f"Python 3.11+ required, found {sys.version_info}"
```

```python
def test_spacy_helper():
    """Test spaCy helper functions (without downloading models)"""
    from data401_nlp.helpers.spacy import list_installed_models
    
    # This should not fail even if no models installed
    models = list_installed_models()
    assert isinstance(models, list)
```

## Integration Test

```python
def test_full_platform_compatibility():
    """Run all platform compatibility tests"""
    print("Testing core imports...")
    test_core_imports()
    
    print("Testing version...")
    test_version()
    
    print("Testing environment detection...")
    test_environment_detection()
    
    print("Testing core dependencies...")
    test_core_dependencies()
    
    print("Checking optional NLP dependencies...")
    test_optional_nlp_dependencies()
    
    print("Testing Python version...")
    test_python_version()
    
    print("Testing spaCy helper...")
    test_spacy_helper()
    
    print("\n✅ All platform compatibility tests passed!")
```

```python
# Run tests when notebook is executed
if __name__ == "__main__":
    test_full_platform_compatibility()
```
